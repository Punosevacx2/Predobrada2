## 7. Predobrada teksta

Tekstualni podaci u svom izvornom obliku često sadrže elemente koji ne doprinose direktno procesu klasifikacije, kao što su velika i mala slova, brojevi, interpunkcijski znakovi i drugi specijalni karakteri. Zbog toga je prije primjene metoda mašinskog učenja neophodno izvršiti predobradu teksta.

U ovom koraku primjenjuju se osnovne tehnike predobrade:
- pretvaranje svih slova u mala slova
- uklanjanje brojeva
- uklanjanje interpunkcije i specijalnih znakova
- uklanjanje suvišnih razmaka

Cilj predobrade je da se tekstualni sadržaj standardizuje i pripremi za naredne faze analize i vektorizacije.

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)                 # uklanjanje brojeva
    text = re.sub(r'[^\w\s]', '', text)            # uklanjanje interpunkcije
    text = re.sub(r'\s+', ' ', text).strip()       # uklanjanje suvišnih razmaka
    return text
```

```python
df["clean_text"] = df["text"].apply(clean_text)
df[["text", "clean_text"]].head(10)
```

## 8. Analiza teksta nakon predobrade

Nakon čišćenja teksta korisno je provjeriti kako predobrada utiče na izgled poruka. Ovaj korak omogućava uvid u to da li su poruke pravilno transformisane i da li je uklonjen neželjeni šum iz podataka.

Takođe se može analizirati dužina poruka nakon predobrade, kako bi se uporedila sa izvornim tekstom.

```python
df["clean_length"] = df["clean_text"].apply(len)
df[["length", "clean_length"]].head()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df["clean_length"], bins=50)
plt.title("Distribucija dužine poruka nakon predobrade")
plt.xlabel("Dužina poruke")
plt.ylabel("Frekvencija")
plt.show()
```

Na osnovu prikazanog histograma može se uočiti raspodjela dužine poruka nakon predobrade. Očekivano je da se prosječna dužina poruka blago smanji, jer su uklonjeni brojevi, interpunkcijski znakovi i drugi elementi koji nisu od suštinskog značaja za semantički sadržaj poruke.

## 9. Priprema ciljnih oznaka

Da bi klasifikacioni model mogao efikasno da obrađuje podatke, potrebno je tekstualne oznake klasa pretvoriti u numerički oblik. U ovom slučaju:
- `ham` će biti mapiran na vrijednost 0
- `spam` će biti mapiran na vrijednost 1

Na ovaj način klasifikacioni zadatak se formuliše kao binarna klasifikacija.

```python
df["label_num"] = df["label"].map({"ham": 0, "spam": 1})
df[["label", "label_num"]].head()
```

```python
df["label_num"].value_counts()
```

## 10. Eksperiment 1: Klasifikacija bez predobrade teksta

Prvi eksperiment služi kao referentna osnova za poređenje. U ovom slučaju koristi se originalni tekst poruka, bez primjene tehnika predobrade. Tekst se transformiše u numerički oblik korišćenjem TF-IDF metode, nakon čega se trenira model logističke regresije.

Rezultati ovog eksperimenta kasnije će biti upoređeni sa rezultatima modela treniranog na prethodno obrađenom tekstu.

```python
X_raw = df["text"]
y = df["label_num"]

X_train_raw, X_test_raw, y_train_raw, y_test_raw = train_test_split(
    X_raw, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
tfidf_raw = TfidfVectorizer()
X_train_raw_tfidf = tfidf_raw.fit_transform(X_train_raw)
X_test_raw_tfidf = tfidf_raw.transform(X_test_raw)
```

```python
model_raw = LogisticRegression(max_iter=1000)
model_raw.fit(X_train_raw_tfidf, y_train_raw)
```

```python
y_pred_raw = model_raw.predict(X_test_raw_tfidf)

accuracy_raw = accuracy_score(y_test_raw, y_pred_raw)
print("Tačnost modela bez predobrade:", accuracy_raw)
```

```python
print(classification_report(y_test_raw, y_pred_raw, target_names=["ham", "spam"]))
```

```python
cm_raw = confusion_matrix(y_test_raw, y_pred_raw)

plt.figure(figsize=(6, 4))
sns.heatmap(cm_raw, annot=True, fmt="d", cmap="Blues")
plt.title("Matrica konfuzije - model bez predobrade")
plt.xlabel("Predikovana klasa")
plt.ylabel("Stvarna klasa")
plt.show()
```

Dobijeni rezultati predstavljaju polaznu tačku za dalje poređenje. Na osnovu vrijednosti tačnosti, preciznosti i odziva može se procijeniti koliko dobro model klasifikuje poruke kada koristi sirovi, neobrađeni tekst.

## 11. Eksperiment 2: Klasifikacija sa predobradom teksta

U drugom eksperimentu koristi se prethodno očišćen tekst iz kolone `clean_text`. Cilj je ispitati da li standardizacija i uklanjanje šuma iz tekstualnih podataka mogu dovesti do poboljšanja performansi klasifikacionog modela.

Za reprezentaciju teksta ponovo se koristi TF-IDF metoda, kako bi poređenje sa prethodnim eksperimentom bilo konzistentno i metodološki ispravno.

```python
X_clean = df["clean_text"]
y = df["label_num"]

X_train_clean, X_test_clean, y_train_clean, y_test_clean = train_test_split(
    X_clean, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
tfidf_clean = TfidfVectorizer()
X_train_clean_tfidf = tfidf_clean.fit_transform(X_train_clean)
X_test_clean_tfidf = tfidf_clean.transform(X_test_clean)
```

```python
model_clean = LogisticRegression(max_iter=1000)
model_clean.fit(X_train_clean_tfidf, y_train_clean)
```

```python
y_pred_clean = model_clean.predict(X_test_clean_tfidf)

accuracy_clean = accuracy_score(y_test_clean, y_pred_clean)
print("Tačnost modela sa predobradom:", accuracy_clean)
```

```python
print(classification_report(y_test_clean, y_pred_clean, target_names=["ham", "spam"]))
```

```python
cm_clean = confusion_matrix(y_test_clean, y_pred_clean)

plt.figure(figsize=(6, 4))
sns.heatmap(cm_clean, annot=True, fmt="d", cmap="Greens")
plt.title("Matrica konfuzije - model sa predobradom")
plt.xlabel("Predikovana klasa")
plt.ylabel("Stvarna klasa")
plt.show()
```

## 12. Poređenje rezultata eksperimenata

U ovom dijelu vrši se direktno poređenje performansi dva klasifikaciona pristupa:
1. model treniran na originalnom tekstu
2. model treniran na prethodno obrađenom tekstu

Poređenje se zasniva prvenstveno na tačnosti klasifikacije, ali se mogu uzeti u obzir i druge metrike kao što su preciznost, odziv i F1 mjera.

```python
results = pd.DataFrame({
    "Eksperiment": ["Bez predobrade", "Sa predobradom"],
    "Tačnost": [accuracy_raw, accuracy_clean]
})

results
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="Eksperiment", y="Tačnost", data=results)
plt.ylim(0.9, 1.0)
plt.title("Poređenje tačnosti modela")
plt.ylabel("Tačnost")
plt.show()
```

## 13. Analiza pogrešno klasifikovanih poruka

Analiza pogrešnih predikcija može pružiti dodatni uvid u ograničenja modela. Posebno je korisno identifikovati poruke koje model klasifikuje pogrešno, jer takvi slučajevi mogu ukazivati na:
- dvosmislene poruke
- specifične izraze karakteristične za spam
- probleme u reprezentaciji teksta

Ovaj korak doprinosi kvalitetnijoj interpretaciji eksperimentalnih rezultata.

```python
results_df = pd.DataFrame({
    "text": X_test_clean.values,
    "real": y_test_clean.values,
    "predicted": y_pred_clean
})

misclassified = results_df[results_df["real"] != results_df["predicted"]]
misclassified.head(10)
```

### Tumačenje rezultata poređenja modela

Na osnovu dobijenih rezultata može se uočiti da je model treniran na originalnom tekstu ostvario tačnost od 0.973094, dok je model treniran na prethodno obrađenom tekstu ostvario tačnost od 0.970404.

Ovi rezultati pokazuju da u posmatranom eksperimentu osnovne tehnike predobrade teksta nisu dovele do poboljšanja performansi klasifikacionog modela. Naprotiv, model bez predobrade ostvario je neznatno bolji rezultat.

Moguće objašnjenje ovakvog ishoda jeste da originalni tekst poruka sadrži određene obrasce, simbole, brojeve i interpunkcijske znakove koji mogu imati diskriminativnu vrijednost prilikom razlikovanja spam i ham poruka. Uklanjanjem takvih elemenata tokom predobrade dio korisnih informacija može biti izgubljen, što se može negativno odraziti na performanse modela.

Takođe, treba uzeti u obzir da je razlika u tačnosti relativno mala, što ukazuje na to da predobrada u ovom slučaju nije imala drastičan uticaj na uspješnost klasifikacije. Ipak, rezultati jasno potvrđuju da efekat predobrade zavisi od prirode skupa podataka i od konkretnih tehnika koje se primjenjuju.

```python
difference = accuracy_raw - accuracy_clean
print(f"Razlika u tačnosti: {difference:.4f}")
print(f"Razlika u procentnim poenima: {difference * 100:.2f}%")
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import nltk

nltk.download("stopwords")

stop_words = set(stopwords.words("english"))
stemmer = PorterStemmer()

def clean_text_advanced(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = text.split()
    tokens = [stemmer.stem(w) for w in tokens if w not in stop_words]
    return " ".join(tokens)

df["clean_text_advanced"] = df["text"].apply(clean_text_advanced)
df[["text", "clean_text_advanced"]].head(10)
```

```python
X_advanced = df["clean_text_advanced"]
y = df["label_num"]

X_train_adv, X_test_adv, y_train_adv, y_test_adv = train_test_split(
    X_advanced, y, test_size=0.2, random_state=42, stratify=y
)

tfidf_adv = TfidfVectorizer()
X_train_adv_tfidf = tfidf_adv.fit_transform(X_train_adv)
X_test_adv_tfidf = tfidf_adv.transform(X_test_adv)

model_adv = LogisticRegression(max_iter=1000, random_state=42)
model_adv.fit(X_train_adv_tfidf, y_train_adv)
```

```python
y_pred_adv = model_adv.predict(X_test_adv_tfidf)
accuracy_adv = accuracy_score(y_test_adv, y_pred_adv)

print("Tačnost modela sa stemmingom i bez stop riječi:", accuracy_adv)
print(classification_report(y_test_adv, y_pred_adv, target_names=["ham", "spam"]))
```

```python
cm_adv = confusion_matrix(y_test_adv, y_pred_adv)

plt.figure(figsize=(6, 4))
sns.heatmap(cm_adv, annot=True, fmt="d", cmap="Purples")
plt.title("Matrica konfuzije - model sa stemmingom i bez stop riječi")
plt.xlabel("Predikovana klasa")
plt.ylabel("Stvarna klasa")
plt.show()
```

```python
results_final = pd.DataFrame({
    "Eksperiment": ["Bez predobrade", "Osnovna predobrada", "Stemming + bez stop riječi"],
    "Tačnost": [accuracy_raw, accuracy_clean, accuracy_adv]
})

print(results_final)

plt.figure(figsize=(9, 5))
sns.barplot(x="Eksperiment", y="Tačnost", data=results_final)
plt.ylim(0.9, 1.0)
plt.title("Poređenje tačnosti sva tri modela")
plt.ylabel("Tačnost")
plt.show()
```
